## Data Cleaning

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

iris_df["Species"]=le.fit_transform(iris_df["Species"])
iris_df.head()
```

This code snippet uses the LabelEncoder from scikit-learn to convert the categorical values in the "Species" column of the iris_df DataFrame into numerical values. The fit_transform method is applied to replace the species names with numerical labels. 

```python
iris_df["Species"].unique()
```


After using LabelEncoder to transform the "Species" column into numerical values, iris_df["Species"].unique() will display the unique numerical labels assigned to the different species. This can be useful to check the encoding result.

```python
x=iris_df.iloc[:,[0,1,2,3]]
x.head()
```

This code extracts the first four columns (features) of the iris_df DataFrame using integer-based indexing with iloc. The resulting DataFrame x contains the values of the features for each entry in the dataset. 

```python
y=iris_df.iloc[:,-1]
y.head()
```

This code extracts the last column of the iris_df DataFrame using integer-based indexing with iloc. The resulting Series y contains the values of the target variable (typically species labels) for each entry in the dataset.

```python
print(x.shape)
print(y.shape)
```

This code prints the shapes of the x and y datasets, providing information about the number of rows and columns in each.

## Model Building

#### Split the data into Traning and Testing Set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=0)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

This code uses scikit-learn's train_test_split function to split the dataset into training and testing sets. It randomly shuffles the data and then allocates a specified portion (in this case, 20%) to the test set, while the rest goes to the training set. The random_state parameter ensures reproducibility.

## Logestic Reggression

```python
from sklearn.linear_model import LogisticRegression 
lr= LogisticRegression()

lr.fit(x_train, y_train) 
print("Logistic regression successfully implemented")

y_pred = lr.predict(x_test)

from sklearn.metrics import confusion_matrix, accuracy_score, classification_report 
cm = confusion_matrix(y_test, y_pred)

print("Confusion Matrix:-")

print(cm)

accuracy = accuracy_score(y_test, y_pred)

print("accuracy is:-", accuracy*100)

print("Classification Report:-")

print(classification_report(y_test,y_pred))
```

This code uses scikit-learn to implement a logistic regression model, make predictions on the test set, and evaluate its performance. 

Split the dataset: The dataset is split into training and testing sets using train_test_split.

Initialize and fit the model: A Logistic Regression model is initialized and trained on the training set using lr.fit(x_train, y_train).

Make predictions: Predictions are made on the test set using lr.predict(x_test).

Evaluate the model:

Confusion Matrix: The confusion matrix is calculated using confusion_matrix, providing a summary of correct and incorrect predictions.
Accuracy: The accuracy is computed using accuracy_score, representing the percentage of correctly predicted instances.
Classification Report: A detailed classification report is generated using classification_report, which includes precision, recall, and F1-score for each class.
